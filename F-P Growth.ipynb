## removing nan values using list over list method

```python
market=[]
for i in range (0,data.shape[0]):
    cus=[]
    for j in data.columns:
        if type(data[j][i])==str:
            cus.append(data[j][i])
            market.append(cus)
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

## Most selling product

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
d={"item name":p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False).head()
```
