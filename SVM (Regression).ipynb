```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:/Data_Analytics/Machine learning/Datasets/placement.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x = data[["cgpa"]]
y = data["package"]
```

```python
sns.scatterplot(x="cgpa",y="package",data=data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=32)
```

```python
from sklearn.svm import SVR
```

```python
sv = SVR(kernel='linear')     #model linearity ko follow krra h isliye linear use kiye h
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
sns.scatterplot(x="cgpa",y="package",data=data)
plt.plot(data["cgpa"],sv.predict(x),color="red")
plt.show()
```
