```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r"D:/Data_Analytics/Machine learning/Datasets/placement.csv")
```

```python
data.head()
```

```python
x= data.iloc[:,:-1]
y= data["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=32)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt =DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
df={
    "criterion":["squared_error","firedown_mse","absolute_error","Poisson"],
   "splitter" :["best","random"], 
   "max_depth":[i for i in range (2,20)]
    }
```

```python
gd= GridSearchCV (DecisionTreeRegressor(),param_grid=df)
```

```python
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_*100
```

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df, n_iter=20)
```

```python
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_*100
```
