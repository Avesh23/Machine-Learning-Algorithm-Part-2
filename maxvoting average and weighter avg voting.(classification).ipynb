```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r"D:/Data_Analytics/Machine learning/Datasets/placement.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
y = data["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=32)
```

```python
    from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

## LinearRegression

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100,lr.score(x_train,y_train)*100
```

## DecisionTreeRegressor

```python
dt =DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)*100,dt.score(x_train,y_train)*100
```

## SVR

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li=[("lr1", LinearRegression()),("sv",SVR()),("dt",DecisionTreeRegressor())]
```

```python
vc= VotingRegressor(li)
vc.fit(x_train,y_train)
```

```python
vc.score(x_test,y_test)*100,vc.score(x_train,y_train)*100
```

##  how it works

```python
prd = {"lr":lr.predict(x_test),"dt":dt.predict(x_test),"sv":sv.predict(x_test),"vc":vc.predict(x_test)}
```
